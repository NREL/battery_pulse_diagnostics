# Examples: Visualizing the dataset

```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd

import sys
sys.path.append("../../")
from utils.data_utils import retime
from utils.data_vis import plot_segment_parts
```

All data files are stored in an HDF5 database format. Individual files can be read using each file key.

```python
hdf = pd.HDFStore('../../data/data_raw.h5', mode='r')
keys = hdf.keys()
keys
```

```python
idx_A = np.nonzero(["_A_" in key for key in keys])
df_A = hdf.get(keys[idx_A[0][0]])
df_A
```

Nissan Leaf data is from 2P-2S modules. The 'Voltage, V' channel has 10 mV resolution instead of 1 mV resolution, as with other data files. There are three additional voltage measurements, one for each parallel group and one for the full cell, with 1 mV resolution.

```python
idx_leaf = np.nonzero(["Leaf" in key for key in keys])
df_leaf = hdf.get(keys[idx_leaf[0][0]])
df_leaf
```

Each file is stored as a dataframe. The data frame contains the raw measurements from the cycler, and the measurement is separated into segments. Segments that repeat (like pulses across the voltage window) are also given segment part numbers.

```python
df_A.plot(x='Total Time, S', y='Voltage, V')
```

Plot one measurement from each cell type.

```python
cell_types = ['/A/', '/B/', '/C/', '/D/']
labels = ['64 Ah NMC-Gr Pouch', '66 Ah LMO-Gr Module', '26 Ah NMC-Gr Prismatic', '2.3 Ah LFP-Gr Cylindrical']
for cell_type, label in zip(cell_types, labels):
    idx = np.nonzero([cell_type in key for key in keys])
    if cell_type == '/B/':
        i = 1
    else:
        i = 0
    df = hdf.get(keys[idx[0][i]])
    plt.plot(df['Total Time, S']/3600, df['Voltage, V'], label=label)

plt.xlabel('Time (hours)')
plt.ylabel('Voltage (V)')
plt.legend()
```

Plot all the measurements.

```python
fig, ax = plt.subplots(1,2, figsize=(10,4))
for key in keys:
    df = hdf.get(key)
    ax[0].plot(df['Total Time, S'], df['Voltage, V'])
    ax[1].plot(df['Total Time, S'], df['SOC'])
hdf.close()
```

Segments with many distinct steps that are worth denoting have many descriptions. Note that these descriptions all share a common root, i.e., 'Equilibration cycle', 'HPPC charge', etc., and can be parsed easily using string parsing functions.

```python
np.unique(df_A['Segment Description'])
```

```python
# Plot all 2C rapid pulse measurements from HPPC in both charge and discharge directions
fig, ax = plt.subplots()

# Mask off data
is_pulse = ['2C rapid pulses' in seg_desc for seg_desc in df_A['Segment Description']]
ax.plot(df_A.loc[is_pulse, 'Segment Time, S'], df_A.loc[is_pulse, 'Voltage, V'])
ax.set_xlabel('Segment Time, S')
ax.set_ylabel('Voltage, V')
ax.set_xlim([0, 16])
```

Note that while the segments start at t=0, the segment parts are not separated, leading to a lot of 'backtracking' lines. There's a nice helper function, `plot_segment_parts`, to handle this. Also, automatically labels your axes by your selected plotting variables. Additional keyword arguments are passed to `plt.plot()`.

```python
# Plot all 2C rapid pulse measurements from HPPC in both charge and discharge directions

import sys
sys.path.append("../../")
from utils.data_vis import plot_segment_parts
fig, ax = plt.subplots()
plot_segment_parts(ax, df_A[is_pulse], x='Segment Time, S', line_style='o')
```

For segments without separate parts, logical filtering is simpler.

```python
fig, ax = plt.subplots()
df = df_A[df_A['Segment Description'] == 'Equilibration cycle charge CC']
ax.plot(df['Segment Time, S'], df['Voltage, V'])
ax.set_xlabel('Segment Time, S')
ax.set_ylabel('Voltage, V')
```

The rate capability cycles have multiple steps to enable the parts to be separated easily for data processing or investigation.

```python
# Plot the 1C cycle, comparing overall and separated
fig, ax = plt.subplots(ncols=2, figsize=(12, 4))
ax = ax.ravel()

# Plot entire cycle
is_cycle = ['1C cycle' in seg_desc for seg_desc in df_A['Segment Description']]
ax[0].plot(df_A.loc[is_cycle, 'Total Time, S'], df_A.loc[is_cycle, 'Voltage, V'])
ax[0].set_xlabel('Total Time, S')
ax[0].set_ylabel('Voltage, V')

# Plot separate components
plot_segment_parts(ax[1], df_A[is_cycle], x='Total Time, S')
ax[1].legend()
```

Combine logical indexing for more complicated filtering. Segment part lines are automatically named with their 'Segment Description'.

```python
# Plot rate capability discharge measurements
fig, ax = plt.subplots()
is_cycle = ['Rate test' in seg_desc and 'discharge CC' in seg_desc in seg_desc for seg_desc in df_A['Segment Description']]
plot_segment_parts(ax, df_A[is_cycle], x='Amp-Hours, AH')
ax.legend()
```

The Pseudo-random pulse test is broken up into smaller 'segments' than we want here. Use the `retime` helper function to fix this, instead of using a one-liner every single time you'd like to redo the segment time for a given data filter. The data overlaps, we'll set a constant offset so we can see all the curves. You can see that the pulses do not affect the charge time.

```python
import sys
sys.path.append("../../")
from utils.data_utils import retime
fig, ax = plt.subplots()

is_Cb2 = ['C/2 cycle charge' in seg_desc for seg_desc in df_A['Segment Description']]
df_ = retime(df_A[is_Cb2])
ax.plot(df_['Segment Time, S'], df_['Voltage, V'], label='C/2')

is_Cb2_PsRP_1 = ['PsRP 1 C/2 diagnostic cycle charge' in seg_desc for seg_desc in df_A['Segment Description']]
df_ = retime(df_A[is_Cb2_PsRP_1])
ax.plot(df_['Segment Time, S'], df_['Voltage, V']+0.2, label='C/2 PsRP 1 (+200 mV)')

is_Cb2_PsRP_2 = ['PsRP 2 C/2 diagnostic cycle charge' in seg_desc for seg_desc in df_A['Segment Description']]
df_ = retime(df_A[is_Cb2_PsRP_2])
ax.plot(df_['Segment Time, S'], df_['Voltage, V']+0.4, label='C/2 PsRP 2 (+400 mV)')

ax.set_xlabel('Segment Time, S')
ax.set_ylabel('Voltage, V')
ax.legend()
```
